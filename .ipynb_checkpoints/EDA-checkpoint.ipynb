```python
#Kepp the categorical variables from de identity dataset
cat_feat = []

for col in train_iden.columns:
    if train_iden[col].dtypes is np.dtype('object'):
        cat_feat.append(col)
    elif train_iden[col].dtypes is np.dtype('int'):
        cat_feat.append(col)
    else:
        pass

print(cat_feat)
```

```python
for feat in cat_feat:
    print(train_iden[feat].value_counts(), '\n')
```

```python
#Kepp the continuos variables from de identity dataset
cont_feat = []

for col in train_iden.columns:
    if train_iden[col].dtypes is np.dtype('float64'):
        cont_feat.append(col)
    else:
        pass

print(cont_feat)
```
